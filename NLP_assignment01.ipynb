```python
import re
import nltk
from nltk import ngrams
import gensim
from docx import Document
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.manifold import TSNE
import numpy as np
nltk.download('averaged_perceptron_tagger_eng')
# Download necessary NLTK data (if not already downloaded)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')  # Check if 'stopwords' is downloaded
except LookupError:
    nltk.download('stopwords')  # Download 'stopwords' if not found

try:
    nltk.data.find('taggers/averaged_perceptron_tagger')  # Check if POS tagger is downloaded
except LookupError:
    nltk.download('averaged_perceptron_tagger')  # Download POS tagger if not found

# Function to clean text (remove URLs, numbers, special characters)
def clean_text(text):
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = re.sub(r'[^a-zA-Z ]+', '', text)  # Remove anything that's not a letter or space
    return text.strip()

# Function to read Word file
def read_word_file(file_path):
    document = Document(file_path)
    articles = []
    for para in document.paragraphs:
        if para.text.strip():
            articles.append(clean_text(para.text))
    return articles

# Function to save cleaned text to a new Word file
def save_to_word(cleaned_texts, output_file):
    if cleaned_texts:
        doc = Document()
        for text in cleaned_texts:
            doc.add_paragraph(text)
        doc.save(output_file)
        print(f"Cleaned text has been saved to {output_file}")
    else:
        print("No articles found. Nothing was saved to the Word file.")

# Function to save POS-tagged text to a new Word file
def save_pos_tagged_text(tagged_texts, output_file):
    if tagged_texts:
        doc = Document()
        for tagged_text in tagged_texts:
            doc.add_paragraph(" ".join([f"{word}/{tag}" for word, tag in tagged_text]))
        doc.save(output_file)
        print(f"POS-tagged text has been saved to {output_file}")
    else:
        print("No POS-tagged text found. Nothing was saved to the Word file.")

# Function to visualize frequent words using WordCloud
def visualize_frequent_words(cleaned_texts):
    if cleaned_texts:
        all_text = ' '.join(cleaned_texts)
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.show()
    else:
        print("No articles found. WordCloud visualization skipped.")

# Function to get top N frequent n-grams
def get_top_ngrams(text, n, top_n=20):
    tokens = nltk.word_tokenize(text)  # Tokenize the text
    ngram_counts = Counter(ngrams(tokens, n))
    return ngram_counts.most_common(top_n)

def prepare_text_for_word2vec(cleaned_texts):
    # Tokenize each cleaned article
    tokenized_texts = [nltk.word_tokenize(text.lower()) for text in cleaned_texts if text]

    # Debugging: Print the first 2 tokenized texts to check structure
    print("Tokenized texts preview:", tokenized_texts[:2])  # Show first 2 tokenized articles
    return tokenized_texts

# Function to train Word2Vec model (using either CBOW or Skip-gram)
def train_word2vec_model(cleaned_texts, sg=1, vector_size=100, window=5, min_count=2):
    # Tokenize the cleaned texts and ensure each article is a list of words
    tokenized_texts = prepare_text_for_word2vec(cleaned_texts)

    if not tokenized_texts:
        raise ValueError("Tokenized texts are empty. Please check the input.")

    # Train Word2Vec model (sg=1 for Skip-gram, sg=0 for CBOW)
    model = gensim.models.Word2Vec(tokenized_texts, vector_size=vector_size, window=window, min_count=min_count, sg=sg)
    return model

# Function to visualize Word2Vec embeddings using t-SNE
def plot_word_embeddings(model, num_words=100):
    words = list(model.wv.index_to_key)[:num_words]

    if not words:
        print("No words available for visualization.")
        return

    word_vectors = [model.wv[word] for word in words]

    if not word_vectors:
        print("No word vectors available for visualization.")
        return

    # Convert word_vectors into a numpy array for t-SNE
    word_vectors = np.array(word_vectors)

    # Use t-SNE for dimensionality reduction to 2D
    tsne = TSNE(n_components=2, random_state=0)
    reduced_vectors = tsne.fit_transform(word_vectors)

    # Plotting
    plt.figure(figsize=(12, 8))
    plt.scatter(reduced_vectors[:, 0], reduced_vectors[:, 1])

    for i, word in enumerate(words):
        plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]), fontsize=12)

    plt.title("Word Embeddings Visualization (t-SNE)")
    plt.show()

# Function to apply POS tagging on the cleaned text
def apply_pos_tagging(cleaned_texts):
    tagged_texts = []
    for text in cleaned_texts:
        tokens = nltk.word_tokenize(text)  # Tokenize the text
        tagged_text = nltk.pos_tag(tokens)  # POS tagging using Penn Treebank tagset
        tagged_texts.append(tagged_text)
    return tagged_texts

# File paths
input_file = "/corpus.docx"  # Update with the correct file path
output_cleaned_file = "cleaned_text.docx"  # Output Word file for cleaned text
output_pos_tagged_file = "pos_tagged_text.docx"  # Output Word file for POS-tagged text

# Main execution
try:
    # Read and clean the document
    articles = read_word_file(input_file)

    # Ensure analysis is performed on cleaned text
    if articles:
        # Save the cleaned text to a new Word file
        save_to_word(articles, output_cleaned_file)

        # Apply POS tagging to the cleaned text
        tagged_articles = apply_pos_tagging(articles)

        # Save the POS-tagged text to a new Word file
        save_pos_tagged_text(tagged_articles, output_pos_tagged_file)

        # Visualize frequent words using WordCloud
        visualize_frequent_words(articles)

        try:
            # Train the Word2Vec model
            model = train_word2vec_model(articles, sg=1)

            # Example: Get vector for a word (Check if the word is in the model's vocabulary)
            word_to_check = 'technologies'  # Change to a word that's in the vocabulary
            if word_to_check in model.wv:
                print(f"Vector for '{word_to_check}':\n", model.wv[word_to_check])
            else:
                print(f"Word '{word_to_check}' not found in the Word2Vec model.")

            # Plot the word embeddings using t-SNE
            plot_word_embeddings(model)

        except Exception as e:
            print(f"An error occurred while training Word2Vec or plotting embeddings: {e}")
    else:
        print("No articles found in the document. Please check the input file.")

except Exception as e:
    print(f"An error occurred: {e}")

from sklearn.feature_extraction.text import TfidfVectorizer

# Function to compute and display TF-IDF scores
def compute_tfidf(cleaned_texts, top_n=10):
    if not cleaned_texts:
        print("No cleaned texts available for TF-IDF computation.")
        return

    # Initialize the TfidfVectorizer
    vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

    # Fit and transform the cleaned texts
    tfidf_matrix = vectorizer.fit_transform(cleaned_texts)

    # Get feature names (terms)
    feature_names = vectorizer.get_feature_names_out()

    # Display top n TF-IDF terms for each document
    print("\nTop TF-IDF terms per document:")
    for idx, doc in enumerate(cleaned_texts):
        tfidf_scores = tfidf_matrix[idx].toarray().flatten()
        top_indices = tfidf_scores.argsort()[-top_n:][::-1]  # Get indices of top n terms
        top_terms = [(feature_names[i], tfidf_scores[i]) for i in top_indices]
        print(f"\nDocument {idx + 1}:")
        for term, score in top_terms:
            print(f"{term}: {score:.4f}")

# Main execution (continued)
try:
    # Read and clean the document
    articles = read_word_file(input_file)

    # Ensure analysis is performed on cleaned text
    if articles:
        # Save the cleaned text to a new Word file
        save_to_word(articles, output_cleaned_file)

        # Apply POS tagging to the cleaned text
        tagged_articles = apply_pos_tagging(articles)

        # Save the POS-tagged text to a new Word file
        save_pos_tagged_text(tagged_articles, output_pos_tagged_file)

        # Visualize frequent words using WordCloud
        visualize_frequent_words(articles)

        # Compute and display TF-IDF scores
        compute_tfidf(articles)

        try:
            # Train the Word2Vec model
            model = train_word2vec_model(articles, sg=1)

            # Example: Get vector for a word (Check if the word is in the model's vocabulary)
            word_to_check = 'technologies'  # Change to a word that's in the vocabulary
            if word_to_check in model.wv:
                print(f"Vector for '{word_to_check}':\n", model.wv[word_to_check])
            else:
                print(f"Word '{word_to_check}' not found in the Word2Vec model.")

            # Plot the word embeddings using t-SNE
            plot_word_embeddings(model)

        except Exception as e:
            print(f"An error occurred while training Word2Vec or plotting embeddings: {e}")
    else:
        print("No articles found in the document. Please check the input file.")

except Exception as e:
    print(f"An error occurred: {e}")

```

```python
import re
import nltk
from nltk import ngrams
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.manifold import TSNE
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('stopwords')

# Function to clean text (remove URLs, numbers, special characters)
def clean_text(text):
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = re.sub(r'[^a-zA-Z ]+', '', text)  # Remove anything that's not a letter or space
    return text.strip()

# Function to read a .txt file
def read_txt_file(file_path):
    articles = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            if line.strip():
                articles.append(clean_text(line.strip()))
    return articles

# Function to save cleaned text to a new .txt file
def save_to_txt(cleaned_texts, output_file):
    if cleaned_texts:
        with open(output_file, 'w', encoding='utf-8') as file:
            for text in cleaned_texts:
                file.write(text + '\n')
        print(f"Cleaned text has been saved to {output_file}")
    else:
        print("No articles found. Nothing was saved to the .txt file.")

# Function to save POS-tagged text to a new .txt file
def save_pos_tagged_text(tagged_texts, output_file):
    if tagged_texts:
        with open(output_file, 'w', encoding='utf-8') as file:
            for tagged_text in tagged_texts:
                tagged_line = " ".join([f"{word}/{tag}" for word, tag in tagged_text])
                file.write(tagged_line + '\n')
        print(f"POS-tagged text has been saved to {output_file}")
    else:
        print("No POS-tagged text found. Nothing was saved to the .txt file.")

# Function to visualize frequent words using WordCloud
def visualize_frequent_words(cleaned_texts):
    if cleaned_texts:
        all_text = ' '.join(cleaned_texts)
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.show()
    else:
        print("No articles found. WordCloud visualization skipped.")

# Function to compute and display TF-IDF scores
def compute_tfidf(cleaned_texts, top_n=10):
    if not cleaned_texts:
        print("No cleaned texts available for TF-IDF computation.")
        return

    # Initialize the TfidfVectorizer
    vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

    # Fit and transform the cleaned texts
    tfidf_matrix = vectorizer.fit_transform(cleaned_texts)

    # Get feature names (terms)
    feature_names = vectorizer.get_feature_names_out()

    # Display top n TF-IDF terms for each document
    print("\nTop TF-IDF terms per document:")
    for idx, doc in enumerate(cleaned_texts):
        tfidf_scores = tfidf_matrix[idx].toarray().flatten()
        top_indices = tfidf_scores.argsort()[-top_n:][::-1]  # Get indices of top n terms
        top_terms = [(feature_names[i], tfidf_scores[i]) for i in top_indices]
        print(f"\nDocument {idx + 1}:")
        for term, score in top_terms:
            print(f"{term}: {score:.4f}")

# File paths
input_file = "corpus.txt"  # Input .txt file
output_cleaned_file = "cleaned_text.txt"  # Output .txt file for cleaned text
output_pos_tagged_file = "pos_tagged_text.txt"  # Output .txt file for POS-tagged text

# Main execution
try:
    # Read and clean the text file
    articles = read_txt_file(input_file)

    # Ensure analysis is performed on cleaned text
    if articles:
        # Save the cleaned text to a new .txt file
        save_to_txt(articles, output_cleaned_file)

        # Apply POS tagging to the cleaned text
        tagged_articles = [nltk.pos_tag(nltk.word_tokenize(text)) for text in articles]

        # Save the POS-tagged text to a new .txt file
        save_pos_tagged_text(tagged_articles, output_pos_tagged_file)

        # Visualize frequent words using WordCloud
        visualize_frequent_words(articles)

        # Compute and display TF-IDF scores
        compute_tfidf(articles)
    else:
        print("No articles found in the text file. Please check the input file.")

except Exception as e:
    print(f"An error occurred: {e}")
```
